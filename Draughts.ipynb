```python
###### " Draughts for Revision "

#WRITING MY BOARD CLASS 

class board():
    
    def __init__(self,grid):
        self.grid=grid

    def display(self,setup):
        print('     A     B     C     D     E     F     G     H\n   _____ _____ _____ _____ _____ _____ _____ _____ \n  |     |     |     |     |     |     |     |     |\n1 |     |  {}  |     |  {}  |     |  {}  |     |  {}  |\n  |_____|_____|_____|_____|_____|_____|_____|_____|\n  |     |     |     |     |     |     |     |     |\n2 |  {}  |     |  {}  |     |  {}  |     |  {}  |     |\n  |_____|_____|_____|_____|_____|_____|_____|_____|\n  |     |     |     |     |     |     |     |     |\n3 |     |  {}  |     |  {}  |     |  {}  |     |  {}  |\n  |_____|_____|_____|_____|_____|_____|_____|_____|\n  |     |     |     |     |     |     |     |     |\n4 |  {}  |     |  {}  |     |  {}  |     |  {}  |     |\n  |_____|_____|_____|_____|_____|_____|_____|_____|\n  |     |     |     |     |     |     |     |     |\n5 |     |  {}  |     |  {}  |     |  {}  |     |  {}  |\n  |_____|_____|_____|_____|_____|_____|_____|_____|\n  |     |     |     |     |     |     |     |     |\n6 |  {}  |     |  {}  |     |  {}  |     |  {}  |     |\n  |_____|_____|_____|_____|_____|_____|_____|_____|\n  |     |     |     |     |     |     |     |     |\n7 |     |  {}  |     |  {}  |     |  {}  |     |  {}  |\n  |_____|_____|_____|_____|_____|_____|_____|_____|\n  |     |     |     |     |     |     |     |     |\n8 |  {}  |     |  {}  |     |  {}  |     |  {}  |     |\n  |_____|_____|_____|_____|_____|_____|_____|_____|\n'.format(setup[0],setup[1],setup[2],setup[3],setup[4],setup[5],setup[6],setup[7],setup[8],setup[9],setup[10],setup[11],setup[12],setup[13],setup[14],setup[15],setup[16],setup[17],setup[18],setup[19],setup[20],setup[21],setup[22],setup[23],setup[24],setup[25],setup[26],setup[27],setup[28],setup[29],setup[30],setup[31]))        
                        
    def select_piece(self,coord,setup):
        Coord=coord.upper()
        if Coord[1]=='0' or Coord[1]=='9':
            return 'No move'
        else:
            current=setup[(self.grid[Coord])-1]
            return current
    
    def reassign(self,Coord,setup,reassignment):
        setup[(self.grid[Coord])-1]=reassignment
        return setup
    
    def move_options(self,coord,setup,piece):
        Coord=coord.upper()
        alpha_num={'A':1,'B':2,'C':3,'D':4,'E':5,'F':6,'G':7,'H':8}
        num_alpha={1:'A',2:'B',3:'C',4:'D',5:'E',6:'F',7:'G',8:'H'}
        
        if piece.lower()=='x':
            num=int(Coord[1])+1
        elif piece.lower()=='o':
            num=int(Coord[1])-1
            
        #LEFT HAND EDGE OF BOARD
            
        if Coord[0]=='A':
            letter1='B'
            coord1=letter1+str(num)
            coord2='No move'
            return coord1,coord2
            
        #RIGHT HAND EDGE OF BOARD
            
        elif Coord[0]=='H':
            letter1='G'
            coord1=letter1+str(num)
            coord2='No move'
            return coord1,coord2
                    
        #MIDDLE OF BOARD
        
        else:
            letter1=num_alpha[(alpha_num[Coord[0]])-1]
            letter2=num_alpha[(alpha_num[Coord[0]])+1]
            coord1=letter1+str(num)
            coord2=letter2+str(num)
                
            return coord1,coord2
            

    def take_test(self,coord,coord_test,piece):      
        x=coord.upper()
        y=coord_test.upper()
        alpha_num={'A':1,'B':2,'C':3,'D':4,'E':5,'F':6,'G':7,'H':8}
        num_alpha={1:'A',2:'B',3:'C',4:'D',5:'E',6:'F',7:'G',8:'H'}

        xx=alpha_num[x[0]]
        yy=alpha_num[y[0]]
       
        if yy<xx:
            check1=num_alpha[yy-1]
        else:
            check1=num_alpha[yy+1]

        if piece.lower()=='x':     
            check2=str(int((y[1]))+1)

        if piece.lower()=='o':
            check2=str(int((y[1]))-1)

        check=check1+check2
        return check            
            
#SETUP IS PIECE LAYOUT ON BOARD
#GRID IS SIMPLY ASIGNING COORDINATES NUMERICALLY IN ORDER

setup = ['x','x','x','x','x','x','x','x','x','x','x','x',' ',' ',' ',' ',' ',' ',' ',' ','o','o','o','o','o','o','o','o','o','o','o','o']           
grid={'B1':1,'D1':2,'F1':3,'H1':4,'A2':5,'C2':6,'E2':7,'G2':8,'B3':9,'D3':10,'F3':11,'H3':12,'A4':13,'C4':14,'E4':15,'G4':16,'B5':17,'D5':18,'F5':19,'H5':20,'A6':21,'C6':22,'E6':23,'G6':24,'B7':25,'D7':26,'F7':27,'H7':28,'A8':29,'C8':30,'E8':31,'G8':32}

#INSTANCIATING MY BOARD CLASS AS 'b'

b=board(grid)

#FUNCTION FOR A SINGLE MOVE

def Turn(coord,piece,test,repeat):

    coord1,coord2=b.move_options(coord,setup,piece)

    if not coord1=='No move':
        piece1=b.select_piece(coord1,setup)
    else:
        piece1='No move'
        
    if not coord2=='No move':
        piece2=b.select_piece(coord2,setup)
    else:
        piece2='No move'
        
    if piece=='X' or piece=='O':
        Move=coord
        Jumped='Nothing'
        Jump='No'
        coord_end1='Unassigned'
        piece_end1='Unassigned'
        coord_end2='Unassigned'
        piece_end2='Unassigned'
        return piece,coord1,piece1,coord2,piece2,coord_end1,piece_end1,coord_end2,piece_end2,Move,Jumped,Jump
        
    if ((piece.lower()=='x' and piece1.lower()=='o') or (piece.lower()=='o' and piece1.lower()=='x')) and (coord1[0]!='A' and coord1[0]!='H') and coord1[1]!='1' and coord1[1]!='8':
        coord_end1=b.take_test(coord,coord1,piece)
        piece_end1=b.select_piece(coord_end1,setup)
    else:
        coord_end1='No take'
        piece_end1='No take'


    if ((piece.lower()=='x' and piece2.lower()=='o') or (piece.lower()=='o' and piece2.lower()=='x')) and (coord2[0]!='A' and coord2[0]!='H') and coord2[1]!='1' and coord2[1]!='8':
        coord_end2=b.take_test(coord,coord2,piece)
        piece_end2=b.select_piece(coord_end2,setup)
    else:
        coord_end2='No take'
        piece_end2='No take'
        
    #MOVE CONDITIONS 
    
    if piece1==' ' and piece2==' ' and repeat=='No':
        if test=='No':
            while True:
                Move=input('Would you like to move to {} or {}?'.format(coord1,coord2)).upper()
                if Move.upper()==coord1.upper() or Move.upper()==coord2.upper():
                    break
                else:
                    print('Please select one of the availble spaces to move to.')
                    
            Jumped='Nothing'
            Jump='No'
            
        elif test=='Yes': 
            Move=coord
            Jumped='Nothing'
            Jump='No'
            
    elif piece_end1==' ' and not piece_end2==' ':
        if test=='No':
            print('You have taken the opposition piece at {} and moved to {}'.format(coord1,coord_end1))
            Move=coord_end1
            Jumped=coord1
            Jump='Yes'
        elif test=='Yes':
            Move=coord_end1
            Jumped=coord1
            Jump='Yes'

    elif not piece_end1==' ' and piece_end2==' ':
        if test=='No':
            print('You have taken the opposition piece at {} and moved to {}'.format(coord2,coord_end2)) 
            Move=coord_end2
            Jumped=coord2
            Jump='Yes'
        elif test=='Yes':
            Move=coord_end2
            Jumped=coord2
            Jump='Yes'
            
    elif piece_end1==' ' and piece_end2==' ':
            if test=='No':
                while True:
                    Move=input('Would you like to jump to {} or {}?'.format(coord_end1,coord_end2)).upper()
                    if Move.upper()==coord_end1.upper() or Move.upper()==coord_end2.upper():
                        break
                    else:
                        print('Please select one of the availble spaces to move to.')
                        
            elif test=='Yes':
                Move=coord
                
            if Move==coord_end1:
                Jumped=coord1
                Jump='Yes'
                
            elif Move==coord_end2:
                Jumped=coord2
                Jump='Yes'

    elif not piece1==' ' and not piece2==' ' and not piece_end1==' ' and not piece_end2==' ' and repeat=='No':
        if test=='No':
            print('You cannot move this piece.')
            Move=coord
            Jumped='Nothing'
            Jump='No'
        elif test=='Yes':
            Move=coord
            Jumped='Nothing'
            Jump='No'
            

    elif piece1==' ' and not piece2==' ' and not piece_end2==' ' and repeat=='No':
        if test=='No':
            print('You have moved to {}'.format(coord1))
            Move=coord1
            Jumped='Nothing'
            Jump='No'
        elif test=='Yes':
            Move=coord1
            Jumped='Nothing'
            Jump='No'

    elif piece2==' ' and not piece1==' ' and not piece_end1==' ' and repeat=='No':
        if test=='No':
            print('You have moved to {}'.format(coord2))
            Move=coord2
            Jumped='Nothing'
            Jump='No'
        elif test=='Yes':
            Move=coord2
            Jumped='Nothing'
            Jump='No'
            
    else:
        Jumped='Nothing'
        Jump='No'
        Move=coord

    return piece,coord1,piece1,coord2,piece2,coord_end1,piece_end1,coord_end2,piece_end2,Move,Jumped,Jump


#FUNCTION FOR A FULL MOVE

def Go(setup,player,coord,test):

    while True:
        try:
            piece=b.select_piece(coord,setup)
        except:
            if test=='Yes':
                Jump='No'
                again='No'
                return setup,Jump,again
            elif test=='No':
                print('Please select one of your pieces.')
                Jump='No'
                again='Yes'
                return setup,Jump,again
            
        if piece!=player.lower() and piece!=player.upper():
            if test=='Yes':
                Jump='No'
                again='No'
                return setup,Jump,again
            elif test=='No':
                print('Please select one of your pieces.')
                Jump='No'
                again='Yes'
                return setup,Jump,again   
        else:
            again='No'
            break

    
    repeat='No'
    
    piece,coord1,piece1,coord2,piece2,coord_end1,piece_end1,coord_end2,piece_end2,Move,Jumped,Jump=Turn(coord,piece,test,repeat)
    Old_coord=coord
    
    if test=='No':
        setup=b.reassign(Old_coord,setup,' ')

    if Jump=='Yes' and test=='No':
        setup=b.reassign(Jumped,setup,' ')
        if piece=='x' and Move[1]=='8':
            setup=b.reassign(Move,setup,'X')
        elif piece=='o' and Move[1]=='1':
            setup=b.reassign(Move,setup,'O')
        else:
            setup=b.reassign(Move,setup,piece)
        b.display(setup)

    elif Jump=='No' and test=='No':
        if piece=='x' and Move[1]=='8':
            setup=b.reassign(Move,setup,'X')
        elif piece=='o' and Move[1]=='1':
            setup=b.reassign(Move,setup,'O')
        else:
            setup=b.reassign(Move,setup,piece)
        b.display(setup)

    test_result=Jump  

    while Jump=='Yes' and test=='No':
        Old_coord=Move
        
        repeat='Yes'
        
        piece,coord1,piece1,coord2,piece2,coord_end1,piece_end1,coord_end2,piece_end2,Move,Jumped,Jump=Turn(Move,piece,test,repeat)
        if Jump=='Yes' and test=='No':
            setup=b.reassign(Old_coord,setup,' ')
            setup=b.reassign(Jumped,setup,' ')
            if piece=='x' and Move[1]=='8':
                setup=b.reassign(Move,setup,'X')
            elif piece=='o' and Move[1]=='1':
                setup=b.reassign(Move,setup,'O')
            else:
                setup=b.reassign(Move,setup,piece)
            b.display(setup)
        
    return setup,test_result,again

def tester(player,setup):

    Jump_options=[]
    for n in ['B1','D1','F1','H1','A2','C2','E2','G2','B3','D3','F3','H3','A4','C4','E4','G4','B5','D5','F5','H5','A6','C6','E6','G6','B7','D7','F7','H7','A8','C8','E8','G8']: 
        setup,test_result,again=Go(setup,player,n,'Yes')
        
        if test_result=='Yes':
            Jump_options.append(n)
    
    l=len(Jump_options)
    w='{} or '
    a=l*w
    a=a[:len(a)-4]
    v='Would you like to jump with '+a+'?'
    
    if l>0:
        if l==1:
            coord=Jump_options[0]
        else:
            while True:
                coord=input(v.format(*Jump_options)).upper()
                if coord in Jump_options:
                    break
                else:
                    print('This piece is not available to move.')
    else:
        while True:
            coord=input('Which piece would you like to move?\nPlease enter a grid reference.').upper()
            c=len(coord)
            if c==2:
                break
    
    return coord

#RUNNING MY MODULES AND FUNCTIONS IN THE RIGHT ORDER TO PLAY GAME

print('Player as "x" starts.')

b.display(setup)

x=0
while True:
    if ('x' in setup or 'X' in setup) and ('o' in setup or 'O' in setup): 
        if x%2==0:
            print('Player "x" to move.')
            again='Yes'
            while again=='Yes':
                coord=tester('x',setup)
                setup,test_result,again=Go(setup,'x',coord,'No')
            x=x+1
        else:
            print('Player "o" to move.')
            again='Yes'
            while again=='Yes':
                coord=tester('o',setup)
                setup,test_result,again=Go(setup,'o',coord,'No')
            x=x+1
    else:
        break
        
print('Thanks for playing.')
```
